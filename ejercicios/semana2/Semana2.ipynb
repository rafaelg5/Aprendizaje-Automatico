```python
X_1 = np.array([[1, i] for (i, j) in data_1])
y_1 = np.array([j for (i, j) in data_1])

X_2 = np.array([[1, i] for (i, j) in data_2])
y_2 = np.array([j for (i, j) in data_2])

X_3 = np.array([[1, i] for (i, j) in data_3])
y_3 = np.array([j for (i, j) in data_3])

X_4 = np.array([[1, i] for (i, j) in data_4])
y_4 = np.array([j for (i, j) in data_4])

X_real = np.array([[1, i] for (i, j) in data_real])
y_real = np.array([j for (i, j) in data_real])


# Crear datos sin peso para gráficas
x_1g = np.array([i for (i, j) in data_1])

x_2g = np.array([i for (i, j) in data_2])

x_3g = np.array([i for (i, j) in data_3])

x_4g = np.array([i for (i, j) in data_4])

x_realg = np.array([i for (i, j) in data_real])
```

## 2. Calcular los modelos de los cuatro dataset proporcionados usando la fórmula:

## W=(X^T X)^(-1)X^(T)y

```python
# W_1
W_1 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_1.T, X_1)), X_1.T), y_1)

# W_2
W_2 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_2.T, X_2)), X_2.T), y_2)

# W_3
W_3 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_3.T, X_3)), X_3.T), y_3)

# W_4
W_4 = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_4.T, X_4)), X_4.T), y_4)
```

## 4. Usando el modelo uno y el dataset data_real, calcular:

```python
# Predicción esperada
def E(f_x):
    return float(sum(f_x)) / float(len(f_x))

exp_pred = E(W_1[1]*x_realg + W_1[0])
y = E(y_real)

# a) Varianza esperada
var = E((W_1[1]*x_realg + W_1[0]) ** 2) - (exp_pred ** 2)

# b) Sesgo esperado
bias = y - exp_pred

print('Varianza:', var)
print('Sesgo:', bias)
```
